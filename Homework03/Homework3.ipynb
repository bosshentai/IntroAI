<h1>Importing Library</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

import seaborn as sns 

sns.set()
```

```python
filereader= pd.read_csv('PS4_GamesSales.csv', encoding='cp1252')
```

<h2> Simple Linear Regresion </h2>

```python
X = filereader.iloc[:,5].values.reshape(-1,1)
Y = filereader.iloc[:,7].values.reshape(-1,1)


linearRegression = LinearRegression() # create object for the class
linearRegression.fit(X,Y) # perform linear regression
Ypred = linearRegression.predict(X) # make predictions

#extrair coeficientes
slope = linearRegression.coef_
intercept = linearRegression.intercept_
```

```python
plt.plot(X, (X * slope + intercept), color= "red")


plt.title('Regresso Linear entre\n')
plt.xlabel('\nEUROPE')
plt.ylabel('REST OF THE WORLD\n')
plt.scatter(X, Y)







```

<h2> Linear Regression using Seaborn </h2>

```python
minXlim = filereader["North America"].min() - 0.1
maxXlim = filereader["North America"].max() + 0.1

graphicLienar = sns.lmplot(
     data=filereader, 
    x="North America", 
    y="Global", 
    ci=None,
    scatter_kws={"color":"blue","alpha":0.3,"lw":1},
    line_kws={"color":"red","alpha":0.5,"lw":2},
    height=6
    )
graphicLienar.set(xlim=(minXlim,maxXlim))

plt.show()
```
